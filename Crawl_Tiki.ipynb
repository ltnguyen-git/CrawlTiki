```python
import requests
import pandas as pd
```

```python
query = 'https://tiki.vn/api/personalish/v1/blocks/listings?limit=40&include=advertisement&aggregations=2&trackity_id=fefc5842-4e9e-3516-6753-d6f4acf320ea&category=1795&page={page}&urlKey=dien-thoai-smartphone'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.0.0'}
```

```python
page = 1
name = []
price = []
quantity_sold = []
rating_average = []
review_count = []
brand_name = []
while True:
  r = requests.get(query.replace('{page}',str(page)),headers = headers)
  print('Get data - page:',page)
  
  if r.status_code != 200:
    print('Hết page')
    break
  
  else:
    data = r.json()['data']
    for item in data:
      name.append(item['name'])
      price.append(item['price'])
      brand_name.append(item['brand_name'])
      try:
        quantity_sold.append(item['quantity_sold']['value'])
      except:
        quantity_sold.append('0')
      
      rating_average.append(item['rating_average'])
      review_count.append(item['review_count'])
    print(f'Page {page} --  Done')
    page +=1
    # if page == 200: break
```

```python
df = pd.DataFrame({'name':name,
                   'brand_name':brand_name,
                   'price':price,
                   'quantity_sold':quantity_sold,
                   'rating_average':rating_average,
                   'review_count':review_count,    
                   }, columns = ['name','brand_name','price','quantity_sold','rating_average','review_count'])
```
